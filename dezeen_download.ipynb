```
import pandas as pd
import numpy as np
import timeit
from tqdm import tqdm
from os.path import basename
import os
import urllib.request
import json
```

```
articles_df = pd.read_pickle('data/articles.pkl')
articles_df
```

```
start = timeit.default_timer() # start the times

for _, row in tqdm(articles_df.iterrows()):
  id = row['id']
  title = row['title']  
  path = 'data/'+str(id)+'/'
  url = row['url']
  images = row['images']
  os.makedirs(path, exist_ok=True)
  info = {
      'id': id,
      'title': title,
      'url': url,
      'images': images
  }
  with open('data/'+str(id)+'/'+str(id)+'.json', 'w') as fp:
    json.dump(info, fp)

  for image in images:
    filename = basename(image)
    if not os.path.exists(path+filename):
      try: urllib.request.urlretrieve(image, path+filename)
      except: break

stop = timeit.default_timer() # stop the timer
print('\nRuntime: {} seconds.'.format(stop-start))
```
